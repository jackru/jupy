```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.describe()
```

```python
props.head()
```

```python
props.info()
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.info()
```

```python
contribs.head()
```

```python
props.prop_name.value_counts().reset_index()
```

```python
prop = props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
```

```python
prop.head()
```

```python
prop.info()
```

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.info()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]) \
.amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
top_donors = merged.groupby(
    ["contributor_firstname", "contributor_lastname"]
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
chart = top_donors.head(5).amount.plot.barh()
chart.set_yticklabels(top_donors.contributor_lastname)
```
